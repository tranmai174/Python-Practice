# Converting datatypes

Check correct datatypes for each variables like datatime , category, int etc.

```python
yt.head()
```

```python
#Changing published date, trending date to datetime data type.

yt.publishedAt = pd.to_datetime(yt.publishedAt)
yt.trending_date = pd.to_datetime(yt.trending_date)
```

```python
#Changing object datatypes to category for categorical variables
#Use a Categorical for efficient storage of an object dtypes column with many repeated values.

yt.channelId = yt.channelId.astype('category')
yt.channelTitle = yt.channelTitle.astype('category')
```

# Add, Rename, Drop, Filter Columns

```python
#To rename column names
yt.rename(columns ={'comment_count':'commentcount'},inplace=True )
```

```python
#To drop column names
yt.drop(["thumbnail_link","ratings_disabled","comments_disabled","description","tags"], inplace=True, axis=1)
```

```python
#Add new column that would contain latest view_count for the particular video
#Since number of views keep on increasing everyday, we will be using only single observation against a video id.
yt.count_max_view = yt.groupby('video_id').view_count.transform(max)
```

```python
#Adding column for year to the dataframe
yt.years = yt.publishedAt.dt.year
```

```python
#Adding column for months to the dataframe
yt.months = yt.publishedAt.dt.month
```

```python
import json
```

```python
#Getting category for the category file
#we need id and title from this file which are under items tag and snippet contain title
with open("/Users/tranngocmai/Documents/DATA SCIENCE/python/project/CA_category_id.json") as f:
    categories = json.load(f)["items"]
cat_dict = {}
category=[]
for cat in categories:
    category.append([cat["id"],cat["snippet"]["title"]])
```

```python
yt_category=pd.DataFrame(category,columns=['categoryId','category']) 
yt_category.head(20)
```

```python
yt_category.categoryId=yt_category.categoryId.astype('int64')
#merge yt and yt_category to get category names
yt_merge = yt.merge(yt_category, on = 'categoryId', how = 'inner')
yt_merge['category']
```

```python
agg_func_count = {'view_count': ['count', 'nunique', 'size']}
yt_merge.groupby(['category']).agg(agg_func_count)
```

```python
agg_func_selection = {'likes': ['first', 'last']}
yt_merge.sort_values(by=['channelTitle'],
            ascending=False).groupby(['channelTitle'
                                        ]).agg(agg_func_selection)
```

# Question 3: Which Type of Videos that Viewers are highly reacting? (Public Response vs Type of Videos)

```python
yt.head()
```

```python
#Create a response percentage column which signifies the public response to a video and engagement b/w viewers and the video
yt_merge["Response_Percentage"] = (yt_merge.likes + yt_merge.dislikes ) * 100 /yt_merge.view_count
yt_merge
```

```python
# Set the width and height of the figure
plt.figure(figsize=(20,10))

# Add title
plt.title("Public Response vs Type of Videos")
plt.xticks(rotation=90)

sns.barplot(x=yt_merge.category, y=yt_merge.Response_Percentage)

# Add label for vertical axis
plt.ylabel("Public Response(in percentage)")
```

Response:

Viewers are highly reacting to Comedy , Music, Education Videos.
